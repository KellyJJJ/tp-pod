Cuando trabajas con un arreglo que contiene un valor np.nan (que representa "Not a Number"), las funciones de Numpy como np.sum(), np.mean(), np.max(), y np.min() pueden comportarse de manera inesperada. Esto es porque np.nan es contagioso: cualquier operación aritmética que lo involucre dará como resultado np.nan.

Para evitar que np.nan influya en los cálculos, puedes usar las funciones de Numpy que ignoran np.nan, como np.nansum(), np.nanmean(), np.nanmax(), y np.nanmin():

# Operación vectorizada vs. Operación loopeada
`numpy` no es solo poderoso por la gama de operaciones que podemos hacer en pocas lineas de código. Sino que por su eficiencia.

Supongamos que tenemos la función $f(x)=10*(x^2/e^x)$. Evaluar la función (en celdas apartes) entre -1 y 1 con 100,000 valores (i) con y (ii) sin un loop `for`. Medir tiempos de cada celda y sacar conclusiones.

```python
import numpy as np
import time

def f(x):
    return 10 * (x**2 / np.exp(x))

x = np.linspace(-1, 1, 1000000)

inicio = time.time()
vector = f(x)
fin = time.time()

print("Tiempo vectorizado:", fin - inicio)
```

```python
inicio = time.time()
resultado = []
for i in x:
    resultado.append(f(i))
fin = time.time()

print("Tiempo con loop:", fin - inicio)
```

Lea el archivo csv (sin `pandas` pero se puede usar `csv`) y quedese con todas las columnas que son numericas. Inspeccionarlo puede hacer el trabajo bastante facil.

Nota: Aunque `numpy` tenga una libreria para leer archivos, en este caso no no es útil por que todas las columas constan de diferentes tipo de dato y ademas el encabezado es `string`.

```python
import csv

with open('Boston.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    data = list(reader)

header = data[0]
rows = data[1:]

indices = []
for i in range(len(header)):
    try:
        float(rows[0][i])
        indices.append(i)
    except ValueError:
        continue

# filas númericas
data = [[float(row[i]) for i in indices] for row in rows]
data = np.array(data)

headerNumerico = [header[i] for i in indices]

print(headerNumerico)
print(data[:5])
```

Una vez procesado el archivo csv, proceda a convertirlo en una matriz de `numpy`.
